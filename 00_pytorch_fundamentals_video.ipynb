### Random tensors

Why random tensors?

Random tensors are important because the way many neural networks learn is that they wich tensors full of random numbers and then adjust those random numbers to better represent the data

`Star with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

Torch random tensors - https://docs.pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create a random tensor of size (3, 4)

random_tensor = torch.rand(2, 4)
random_tensor
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 214)) # height, width, colour channel
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

```python
torch.rand(3, 3)
```

### Tensor datatypes

**Note:** Tensor datatypes is one of the 3 big errors you'll run into with PyTorch and deep learning:

1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None, # what datatype is the tensor (e.g float32 or float16)
                               device=None, # what device is your tensor on
                               requires_grad=False) # whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor * float_32_tensor
```

```python
int_32_tensor = torch.tensor([3, 6, 9], dtype=torch.long)
int_32_tensor
```

```python
float_32_tensor * int_32_tensor
```

### Getting information from tensors (tensor attribute)

1. Tensors not right datatype - to do get datatype from a tensor, can use `tensor.dtype`
2. Tensors not right shape - to get shape from a tensor, can use `tensor.shape`
3. Tensors not on the right device - to get device from a tensor, can use `tensor.device`

```python
# Create a tensor
some_tensor = torch.rand(3, 4)
some_tensor
```

```python
some_tensor.size, some_tensor.shape
```

```python
some_tensor.size()
```

```python
# Find out details about some tensor
print(some_tensor)
print(f"Datatype of tensor: {some_tensor.dtype}")
print(f"Shape of tensor: {some_tensor.shape}")
print(f"Device tensor is on: {some_tensor.device}")
```

### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

```python
# Element wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
tensor
```
