### 2. Create a random tensor with shape (7, 7)

````python
import torch

# tensor between 0 and 1
x = torch.rand(7, 7)
print(x)

# tensor with others values
y = torch.round(torch.rand(3, 4) * 10)
print(y)
````

### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape (1, 7) (hint: you may have to transpose the second tensor).

````python
y = torch.rand(1, 7)

# multiplication
# multiplication = torch.matmul(x, y) - generate error
multiplication = torch.matmul(x, torch.t(y)) # Success
multiplication_2 = torch.matmul(x, y.T) # Success - other form

print(multiplication)
print("-----------------------------------")
print(multiplication_2)
````
